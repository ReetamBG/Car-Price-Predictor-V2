```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("./data/cardekho_dataset.csv").iloc[:, 1:]
df.head()
```

```python
df = df[["car_name", "vehicle_age", "km_driven", "mileage", "engine", "max_power", "seats", "selling_price"]]
df.head()
```

```python
car_name_freq = df["car_name"].value_counts()
car_name_freq
```

```python
df["car_name_freq"] = df["car_name"].apply(lambda name: car_name_freq[name])
df.head()
```

```python
df =  df.drop(columns="car_name")
```

```python
df.to_csv("./data/data_preprocessed.csv", index=False)
```
